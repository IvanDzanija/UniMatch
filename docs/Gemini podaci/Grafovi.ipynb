```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load datasets
universities_data = pd.read_excel("Universities_DataFull_Gemini.xlsx")
prices_data = pd.read_excel("pricesComplete.xlsx")

# Standardize column names
universities_data.rename(columns={"Country": "country"}, inplace=True)

# Ensure country names are standardized (trim spaces, convert to lowercase for consistency)
universities_data["country"] = universities_data["country"].str.strip().str.lower()
prices_data["country"] = prices_data["country"].str.strip().str.lower()

# Merge datasets on country
merged_df = pd.merge(universities_data, prices_data, on="country", how="inner")

# Convert tuition columns to numeric
merged_df["Tuition_LLM"] = pd.to_numeric(merged_df["Tuition"], errors="coerce")
merged_df["Tuition"] = pd.to_numeric(merged_df["tuition"], errors="coerce")

# Drop rows with missing tuition values
merged_df = merged_df.dropna(subset=["Tuition_LLM", "Tuition"])

# Train Linear Regression Model
model = LinearRegression()
X = merged_df["Tuition_LLM"].values.reshape(-1, 1)  # Reshape for sklearn
y = merged_df["Tuition"].values
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Calculate R² Score
r2 = r2_score(y, y_pred)
print(f"R² Score: {r2:.4f}")

# Plot
plt.figure(figsize=(8, 6))
sns.regplot(data=merged_df, x="Tuition_LLM", y="Tuition", 
            ci=None, line_kws={"color": "red"})

plt.xlabel("Tuition_Gemini")
plt.ylabel("Tuition")
plt.title(f"Tuition Cost (R² = {r2:.4f})")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load datasets
universities_data = pd.read_excel("Universities_Data.xlsx")
prices_data = pd.read_excel("pricesComplete.xlsx")

# Standardize column names
universities_data.rename(columns={"Country": "country"}, inplace=True)

# Ensure country names are standardized (trim spaces, convert to lowercase for consistency)
universities_data["country"] = universities_data["country"].str.strip().str.lower()
prices_data["country"] = prices_data["country"].str.strip().str.lower()

# Merge datasets on country
merged_df = pd.merge(universities_data, prices_data, on="country", how="inner")

# Convert tuition columns to numeric
merged_df["Tuition_LLM"] = pd.to_numeric(merged_df["Tuition"], errors="coerce")
merged_df["Tuition"] = pd.to_numeric(merged_df["tuition"], errors="coerce")

# Drop rows with missing tuition values
merged_df = merged_df.dropna(subset=["Tuition_LLM", "Tuition"])

# Train Linear Regression Model
model = LinearRegression()
X = merged_df["Tuition_LLM"].values.reshape(-1, 1)  # Reshape for sklearn
y = merged_df["Tuition"].values
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Calculate R² Score
r2 = r2_score(y, y_pred)
print(f"R² Score: {r2:.4f}")

# Plot
plt.figure(figsize=(8, 6))
sns.regplot(data=merged_df, x="Tuition_LLM", y="Tuition", 
            ci=None, line_kws={"color": "red"})

plt.xlabel("Tuition_LLM")
plt.ylabel("Tuition")
plt.title(f"Tuition Cost (R² = {r2:.4f})")

plt.show()
```
